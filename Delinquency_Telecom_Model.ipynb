```python
# after abserving the above results, I am removing the features which have -ve corelation with label 
df = df.drop('fr_da_rech90', axis = 1)
df = df.drop('medianmarechprebal30', axis = 1)
df = df.drop('aon', axis = 1)
df = df.drop('fr_da_rech30', axis = 1)
df = df.drop('pcircle', axis = 1)
df = df.drop('maxamnt_loans30', axis = 1)
df = df.drop('Unnamed: 0', axis = 1)
```

```python
# Checking the shape of the dataset after removing the 4 columns
df.shape
```

```python
# assigning all columns to 'X' except the 'label' column
# Label column is assigned to 'y'
X = df.drop('label', axis = 1).values
y = df['label'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
# Scaling and transforming
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# importing needful librabries to create a model.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
# checking the shape of the X_train
X_train.shape
```

```python
# checking the shape of the X_test
X_test.shape
```

```python
# Creating the model using ANN
model = Sequential()

model.add(Dense(29, activation = 'relu' ))
model.add(Dropout(0.25))
model.add(Dense(14, activation = 'relu' ))
model.add(Dropout(0.25))
model.add(Dense(7, activation = 'relu' ))
model.add(Dropout(0.25))

model.add(Dense(1, activation = 'sigmoid' ))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
# Here I am using 'EarlyStop' to get the accuracy without any time delay
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor ='val_loss', mode = 'min', verbose = 1, patience = 25)
```

```python
# Fitting the mdoel
model.fit(x = X_train, y = y_train, epochs = 600, validation_data = (X_test,y_test), callbacks = [early_stop])
```

```python
# loading the losses into pandas dataframe for visulization purpose
losses = pd.DataFrame(model.history.history)
losses
```

```python
losses.plot()
```

```python
# evaluating the model accuracy
predictions = model.predict_classes(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, predictions))
```
